```python
import requests
import json
from datetime import datetime

# --- YOUR CREDENTIALS ---
# Paste your master API key between the quotes below:
NOTION_TOKEN = "YOUR_NOTION_TOKEN"

# Your specific Database ID is already injected:
DATABASE_ID = "YOUR_DATABASE_ID"

# --- THE DATA YOU WANT TO LOG ---
activity_name = "Python Coding & API Practice"
hours_spent = 2.5
current_date = datetime.now().strftime("%Y-%m-%d")

# --- THE API SETUP ---
url = "https://api.notion.com/v1/pages"

headers = {
    "Authorization": f"Bearer {NOTION_TOKEN}",
    "Content-Type": "application/json",
    "Notion-Version": "2022-06-28"
}

# --- THE JSON PAYLOAD (Formatting for Notion) ---
data = {
    "parent": {"database_id": DATABASE_ID},
    "properties": {
        "Activity": {
            "title": [
                {
                    "text": {
                        "content": activity_name
                    }
                }
            ]
        },
        "Hours": {
            "number": hours_spent
        },
        "Date": {
            "date": {
                "start": current_date
            }
        }
    }
}

# --- SEND THE REQUEST ---
response = requests.post(url, headers=headers, data=json.dumps(data))

if response.status_code == 200:
    print("SUCCESS! Check your Notion Database.")
else:
    print(f"FAILED. Error Code: {response.status_code}")
    print(response.text)
```

```python
# --- 0. GRAB THE NEW PACKAGES ---

from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# --- 1. LOAD THE DOCUMENT ---
print("Loading the PDF...")
loader = PyPDFLoader("research_paper.pdf")
pages = loader.load()
print(f"Success! Loaded {len(pages)} pages.")

# --- 2. SLICE THE DATA ---
print("Slicing data into readable chunks...")
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=400
)
chunks = text_splitter.split_documents(pages)

print(f"Data successfully sliced into {len(chunks)} searchable chunks!")
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma

# --- 1. AUTHENTICATE THE EMBEDDING ENGINE ---
print("Waking up Google's Embedding Engine...")
# Paste your API key here one more time to authenticate this specific tool
GEMINI_API_KEY = "YOUR_GEMINI_API_KEY"

embeddings = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-2-preview",
    google_api_key=GEMINI_API_KEY
)

# --- 2. BUILD THE VECTOR DATABASE ---
print("Converting your 109 chunks into math vectors (This might take 30-60 seconds)...")

# This creates the database and saves it in a folder called "chroma_db"
vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="./chroma_db"
)

print("SUCCESS! Your Vector Database is fully built and ready to be searched.")
```

```python
# --- 0. GRAB THE CLASSIC PACKAGE ---
!pip install -q -U langchain-classic

from langchain_classic.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI

# --- 1. WAKE UP THE REASONING ENGINE ---
print("Waking up the Gemini Brain...")
# Paste your API key here
GEMINI_API_KEY = "YOUR_GEMINI_API_KEY"

# We use temperature=0.2 so the AI gives factual answers instead of creative ones
llm = ChatGoogleGenerativeAI(
    model="gemini-3.5-flash",
    google_api_key=GEMINI_API_KEY,
    temperature=0.2
)

# --- 2. CONNECT BRAIN TO MEMORY ---
print("Connecting Brain to the Vector Database...")
# We tell the database to retrieve the top 3 most relevant chunks for any question
retriever = vector_db.as_retriever(search_kwargs={"k": 3})

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)

# --- 3. ASK THE RESEARCH ENGINE ---
question = "What specific chemical reagents are required to precipitate Group II A cations, and what color are the precipitates?"

print(f"\nQUESTION: {question}")
print("Searching database and generating answer...\n")

response = qa_chain.invoke({"query": question})

print("--- AI ANSWER ---")
print(response["result"])
```

```python
import requests
import json
from datetime import datetime

# --- 1. GRAB YOUR EXISTING CREDENTIALS ---
NOTION_TOKEN = "YOUR_NOTION_TOKEN"
DATABASE_ID = "YOUR_DATABASE_ID"

# --- 2. DYNAMICALLY LOG THE AI'S WORK ---
# We are pulling the exact 'question' variable from the cell above!
activity_name = f"AI Research Query: {question}"
hours_spent = 0.5
current_date = datetime.now().strftime("%Y-%m-%d")

# --- 3. NOTION API SETUP ---
url = "https://api.notion.com/v1/pages"
headers = {
    "Authorization": f"Bearer {NOTION_TOKEN}",
    "Content-Type": "application/json",
    "Notion-Version": "2022-06-28"
}

data = {
    "parent": {"database_id": DATABASE_ID},
    "properties": {
        "Activity": {
            "title": [{"text": {"content": activity_name}}]
        },
        "Hours": {"number": hours_spent},
        "Date": {"date": {"start": current_date}}
    }
}

# --- 4. FIRE THE LOGGER ---
print(f"Pushing to Notion: '{activity_name}'...")
response = requests.post(url, headers=headers, data=json.dumps(data))

if response.status_code == 200:
    print("SUCCESS! Your AI research task was just logged in your database.")
else:
    print(f"FAILED. Error Code: {response.status_code}")
```
